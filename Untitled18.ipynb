```python
import pandas as pd
import numpy as np
import nltk
from matplotlib import pyplot
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
#Loading Dataset
data = pd.read_csv('/content/airport_reviews.csv')
```

```python
data.head()
```

```python
#Tokenization
from nltk.tokenize import word_tokenize

# Applying word_tokenize to the 'content' column of the DataFrame
tokens = data['content'].apply(word_tokenize)

# Display the first few rows of the tokens
print(tokens.head())
```

```python
#Stemming
from nltk import PorterStemmer
porter = PorterStemmer()
stemmed_tokens = tokens.apply(lambda x: [porter.stem(t) for t in x])
print (stemmed_tokens.head())
```

```python
#Removing Stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
data['stemmed_tokens'] = stemmed_tokens # Add stemmed_tokens as a new column
data['cleaned_stemmed_tokens'] = data['stemmed_tokens'].apply(lambda x: [t for t in x if t not in stop_words])
```

```python
data['cleaned_text'] = data['cleaned_stemmed_tokens'].apply(lambda x: ' '.join(x))
```

```python
print(data['cleaned_text'])
```

```python
#Implementing LDA
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer()# Fit and transform the processed titles
count_data = count_vectorizer.fit_transform(data['cleaned_text'])
count_data
vectorizer = CountVectorizer(max_df=0.9, min_df=5, stop_words='english')
dtm = vectorizer.fit_transform(data['cleaned_text'])
lda = LatentDirichletAllocation(n_components=10, random_state=2025)
lda.fit(dtm)
for i, topic in enumerate(lda.components_):
    print(f"Topic {i+1}:")
    print([vectorizer.get_feature_names_out()[index] for index in topic.argsort()[-10:]])
    print()
```

```python
import numpy as np
terms = count_vectorizer.get_feature_names_out()
# Count the popularity of words
total_counts = np.zeros(len(terms))
for t in count_data:
    total_counts+=t.toarray()[0]
count_dict = (zip(terms, total_counts))
#Take the top 50 words
count_dict = sorted(count_dict, key=lambda x:x[1], reverse=True)[0:50]
words = [w[0] for w in count_dict]
counts = [w[1] for w in count_dict]
x_pos = np.arange(len(words))
pyplot.figure(2, figsize=(11, 3))
pyplot.subplot(title='50 most common words')
pyplot.bar(words, counts)
pyplot.xticks(x_pos, words, rotation=90)
pyplot.xlabel('words')
pyplot.ylabel('counts')
pyplot.show()
```
